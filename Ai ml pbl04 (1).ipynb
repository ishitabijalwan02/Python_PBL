```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("agri_dataset.csv")

# Display dataset
df
```

```python
# Structure of dataset
df.info()

# Statistical summary
df.describe()

# Missing values count
df.isnull().sum()
```

```python
# Drop 'region' column
df.drop(columns=['Region'], inplace=True)

df
```

```python
df.columns
```

```python
# Handle missing values 

df['Rainfall_mm'] = df['Rainfall_mm'].fillna(df['Rainfall_mm'].mean())
df['Temperature_C'] = df['Temperature_C'].fillna(df['Temperature_C'].mean())
df['Soil_Moisture_pct'] = df['Soil_Moisture_pct'].fillna(df['Soil_Moisture_pct'].mean())
df['Soil_pH'] = df['Soil_pH'].fillna(df['Soil_pH'].median())

df['Nitrogen'] = df['Nitrogen'].fillna(df['Nitrogen'].mean())
df['Phosphorus'] = df['Phosphorus'].fillna(df['Phosphorus'].mean())
df['Potassium'] = df['Potassium'].fillna(df['Potassium'].mean())

df['Irrigation_pct'] = df['Irrigation_pct'].fillna(df['Irrigation_pct'].mean())
df['Crop_Area_ha'] = df['Crop_Area_ha'].fillna(df['Crop_Area_ha'].mean())

df['Fertilizer_kg_per_ha'] = df['Fertilizer_kg_per_ha'].fillna(df['Fertilizer_kg_per_ha'].mean())
df['Pesticide_kg_per_ha'] = df['Pesticide_kg_per_ha'].fillna(df['Pesticide_kg_per_ha'].mean())

df['Sunlight_hours'] = df['Sunlight_hours'].fillna(df['Sunlight_hours'].mean())
df['Humidity_pct'] = df['Humidity_pct'].fillna(df['Humidity_pct'].mean())

df['Altitude_m'] = df['Altitude_m'].fillna(df['Altitude_m'].mean())
df['Wind_kmh'] = df['Wind_kmh'].fillna(df['Wind_kmh'].mean())
df['Groundwater_m'] = df['Groundwater_m'].fillna(df['Groundwater_m'].mean())

df['Crop_Diversity_Index'] = df['Crop_Diversity_Index'].fillna(df['Crop_Diversity_Index'].mean())
df['Mechanization_Score'] = df['Mechanization_Score'].fillna(df['Mechanization_Score'].mean())
df['Yield_Variance'] = df['Yield_Variance'].fillna(df['Yield_Variance'].mean())

df['Avg_Yield_ton_per_ha'] = df['Avg_Yield_ton_per_ha'].fillna(df['Avg_Yield_ton_per_ha'].mean())

df
```

```python
# Remove duplicate rows

df.drop_duplicates(inplace=True)

df
```

```python
# Removing outliers from Rainfall

Q1 = df['Rainfall_mm'].quantile(0.25)
Q3 = df['Rainfall_mm'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Rainfall_mm'] >= lower) & (df['Rainfall_mm'] <= upper)]
```

```python
# Removing outliers from Temperature

Q1 = df['Temperature_C'].quantile(0.25)
Q3 = df['Temperature_C'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Temperature_C'] >= lower) & (df['Temperature_C'] <= upper)]
```

```python
# Removing outliers from Soil Moisture

Q1 = df['Soil_Moisture_pct'].quantile(0.25)
Q3 = df['Soil_Moisture_pct'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Soil_Moisture_pct'] >= lower) & (df['Soil_Moisture_pct'] <= upper)]
```

```python
df
```

```python
# Converting all values to positive

df = df.abs()

df
```

```python
# Histogram of Yield

plt.hist(df['Avg_Yield_ton_per_ha'], edgecolor='black')
plt.title('Histogram of Yield')
plt.xlabel('Yield')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Rainfall

plt.boxplot(df['Rainfall_mm'])
plt.title('Boxplot of Rainfall')
plt.ylabel('Rainfall')
plt.show()
```

```python
# Scatter plot between Rainfall and Yield

plt.scatter(df['Rainfall_mm'], df['Avg_Yield_ton_per_ha'], edgecolors='black')
plt.title('Rainfall vs Yield')
plt.xlabel('Rainfall')
plt.ylabel('Yield')
plt.show()
```

```python
# Line plot of Temperature

plt.plot(df['Temperature_C'])
plt.title('Temperature Trend')
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.show()
```
